#### Drop Feature which has negative corelation 

```python
df = df.drop(['unique_id','no_of_children','employment','housing_status','marital_status','income_level','sex'],axis=1)
```

### Independent Vairiable

```python
x=df.drop('h1n1_vaccine',axis=1)
```

### Dependent Variable

```python
y=df[['h1n1_vaccine']]
```

```python
x.head()
```

```python
y.head()
```

Import Library to Split Data

```python
from sklearn.model_selection import train_test_split
```

### Spliting of Data

Here we splitting the data into 70% To 30% so 70% is our train model data and 30 % is our test data on that base
we will check our test accuray

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

Print the spliting Data

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Import Logistic Regression Library

```python
from sklearn.linear_model import LogisticRegression
```

Object Creation

```python
model=LogisticRegression()
```

Train the Data

```python
model.fit(x_train,y_train)
```

Predict the Model

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

Accuracy of Train Data model

```python
model.score(x_train,y_train)*100
```

Accuracy of Test Data model

```python
accuracy_score(y_test,y_pred)*100
```

Import Library For Confusion Metrix

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
accuracy_score(y_test,y_pred)*100
```

Create Confusion Metrix

```python
cf=confusion_matrix(y_test,y_pred)
```

Print Cofusion Metrix

```python
cf
```

### Heat Map of Confusion Metrix

```python
plt.figure(figsize=(12,5))
sns.heatmap(cf,annot=True,fmt='g',cmap='Blues')
plt.title('Confusion Matrix',fontsize=20)
plt.show()
```

#Clasification Report

```python
print(classification_report(y_test,y_pred))
```
