## Data Preprocessing

### Handling Missing values

Fill the missing values through median of the data

```python
df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].median(),inplace=True)
```

drop the data because not use of this feature

```python
df.drop(['Unnamed: 0','id'],axis=1,inplace=True)
```

Import Label encoder library to preprocessing the data 

```python
from sklearn.preprocessing import LabelEncoder
```

object Create

```python
le = LabelEncoder()
```

this change the data type and fill the null or string values

```python
df['Gender'] = le.fit_transform(df['Gender'])
df['Customer Type'] = le.fit_transform(df['Customer Type'])
df['Type of Travel'] = le.fit_transform(df['Type of Travel'])
df['Class'] = le.fit_transform(df['Class'])
df['satisfaction'] = le.fit_transform(df['satisfaction'])
```

```python
X = df.drop('satisfaction',axis=1)
```

```python
y = df[['satisfaction']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion='entropy',max_depth=3)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
#training accuracy
model.score(X_train,y_train)*100
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
cf=confusion_matrix(y_test,y_pred)
cf
```

```python
plt.figure(figsize=(12,5))
sns.heatmap(cf,annot=True,fmt='g',cmap='Blues')
plt.title('Confusion Matrix',fontsize=20)
plt.show()
```
