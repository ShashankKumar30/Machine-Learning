```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.cluster import KMeans
```

```python
std=StandardScaler()
```

```python
df_std=df.drop(['Cust_ID','Gender'],axis=1)
```

```python
df_std=std.fit_transform(df_std)
```

```python
df_std
```

```python
df_std=pd.DataFrame(df_std)
```

```python
df_std
```

```python
kmeans=KMeans(n_clusters=4)
```

```python
kmeans.fit(df_std)
```

```python
#segmentation of clustring or grouping
kmeans.labels_
```

```python
df['Cluster']=kmeans.labels_
```
